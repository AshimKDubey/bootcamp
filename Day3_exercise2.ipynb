```python
import pandas as pd
import numpy as np

import bokeh.plotting
import bokeh.io

bokeh.io.output_notebook()
```

```python
df = pd.read_csv('data/collins_switch.csv', comment='#')
```

```python
df
```

```python
x = '[IPTG] (mM)'
y = 'normalized GFP expression (a.u.)'

p = bokeh.plotting.figure(
    frame_width = 300,
    frame_height = 300,
    x_axis_label = x,
    y_axis_label = y,
    x_range = [0,12],
    y_range = [0,1],
)

source = bokeh.models.ColumnDataSource(df) #notneeded

p.circle(
    source=df,#or put df here directly
    x=x,
    y=y,
)

bokeh.io.show(p)
```

```python
x = '[IPTG] (mM)'
y = 'normalized GFP expression (a.u.)'

p = bokeh.plotting.figure(
    frame_width = 300,
    frame_height = 300,
    x_axis_label = x,
    y_axis_label = y,
    x_range = [0.0001,100],
    y_range = [0,1],
    x_axis_type='log',
)

source = bokeh.models.ColumnDataSource(df) #notneeded

p.circle(
    source=df,#or put df here directly
    x=x,
    y=y,
)

bokeh.io.show(p)
```

```python
df['error_low']=df['normalized GFP expression (a.u.)']-1.96*df['sem']
df['error_high']=df['normalized GFP expression (a.u.)']+1.96*df['sem']

df.head()
```

```python
x = '[IPTG] (mM)'
y = 'normalized GFP expression (a.u.)'
error_high = 'error_high'
error_low = 'error_low'

p = bokeh.plotting.figure(
    frame_width = 300,
    frame_height = 300,
    x_axis_label = x,
    y_axis_label = y,
    x_range = [0.0001,100],
    y_range = [0,1],
    x_axis_type='log',
)

source = bokeh.models.ColumnDataSource(df) #notneeded

p.circle(
    source=df,#or put df here directly
    x=x,
    y=y,
)

p.segment(
    source=df,
    x0=x,
    x1=x,
    y1 = 'error_high',
    y0 = 'error_low',
    line_width=2
)

bokeh.io.show(p)
```
