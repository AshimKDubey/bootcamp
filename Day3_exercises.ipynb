```python
df.groupby('ID')['impact force (mN)'].agg(['std']).reset_index()
```

```python
def coeff_of_var(data):
    return np.mean(data)/np.std(data)
# cols = df.columns[(df.columns == 'impact force (mN)') ]
df.groupby('ID')['impact force (mN)'].agg(coeff_of_var).reset_index()
```

```python
df.groupby('ID')[['impact force (mN)', 'adhesive force (mN)']].agg(coeff_of_var).reset_index()
```

```python
df.groupby('ID')[['impact force (mN)', 'adhesive force (mN)']].agg(['mean','median','std',coeff_of_var]).reset_index()
```

```python
ll = df.groupby('ID')[['impact force (mN)', 'adhesive force (mN)']].agg(['mean','median','std',coeff_of_var]).reset_index()
df2 = pd.melt(ll, value_name='Value', var_name=['Force', 'Method'], id_vars='ID')
df2
```
