```python
import sympy
```

```python
n_max =1000

primes = []

for x in range(n_max+1):
    if sympy.isprime(x):
        primes.append(x)

primes[1:10]
```

```python
# List comprehension! - faster due to memory allocation, but mostly awesome for readability!

primes = [x for x in range(n_max+1) if sympy.isprime(x)]

primes[1:10]
```

```python
def coeff_of_var(data):
    return np.mean(data)/np.std(data)
```

```python
# Groupby objects can be indexed, and analysis can be chained!
cols = df.columns[(df.columns!='gender') & (df.columns != 'insomnia')]
df.groupby('insomnia')[cols].agg(coeff_of_var).reset_index()
```

```python
df.groupby(['gender', 'insomnia'])['percent correct'].mean().reset_index()
```

```python
def insom_scale(x):
    if x<=16:
        return 'insomniac'
    elif x<=25:
        return 'restless'
    else:
        return 'restful'

df['insom_scale'] = df['sci'].apply(insom_scale)
```

```python
p = bokeh.plotting.figure(
    frame_width = 300,
    frame_height = 300,
    x_axis_label = 'confidence when correct',
    y_axis_label = 'confidence when incorrect',
    x_range = [-5,105],
    y_range = [-5,105],
)
```

```python
source = bokeh.models.ColumnDataSource(df) #notneeded
p.circle(
    source=source,#or put df here directly
    x='confidence when correct',
    y='confidence when incorrect',
)
```

```python
bokeh.io.show(p)
```

```python
x = 'confidence when correct'
y = 'confidence when incorrect'

p = bokeh.plotting.figure(
    frame_width = 300,
    frame_height = 300,
    x_axis_label = x,
    y_axis_label = y,
    x_range = [-5,105],
    y_range = [-5,105],
)

source = bokeh.models.ColumnDataSource(df) #notneeded

p.circle(
    source=source,#or put df here directly
    x=x,
    y=y,
)

bokeh.io.show(p)
```

```python
x = 'confidence when correct'
y = 'confidence when incorrect'

df_insomnia=df.loc[df['insomnia'], :]
df_normal = df.loc[~df['insomnia'], :]

p = bokeh.plotting.figure(
    frame_width = 300,
    frame_height = 300,
    x_axis_label = x,
    y_axis_label = y,
    x_range = [-5,105],
    y_range = [-5,105],
)

source = bokeh.models.ColumnDataSource(df) #notneeded

p.circle(
    source=df_normal,
    x=x,
    y=y,
    #color = 'blue',
    legend_label='normal'
)

p.circle(
    source=df_insomnia,
    x=x,
    y=y,
    color = 'orange',
    legend_label='insomniacs'
)

p.legend.location = 'top_left'
p.legend.click_policy = 'mute'

bokeh.io.show(p)
```

```python
x = 'confidence when correct'
y = 'confidence when incorrect'

df_insomnia=df.loc[df['insomnia'], :]
df_normal = df.loc[~df['insomnia'], :]

p = bokeh.plotting.figure(
    frame_width = 300,
    frame_height = 300,
    x_axis_label = x,
    y_axis_label = y,
    x_range = [-5,105],
    y_range = [-5,105],
)

source = bokeh.models.ColumnDataSource(df) #notneeded

p.line(
    x=[-5, 105],
    y=[-5, 105],
    color='gray'
)
p.circle(
    source=df_normal,
    x=x,
    y=y,
    #color = 'blue',
    legend_label='normal'
)

p.circle(
    source=df_insomnia,
    x=x,
    y=y,
    color = 'orange',
    legend_label='insomniacs'
)

p.legend.location = 'top_left'
p.legend.click_policy = 'mute'

bokeh.io.show(p)
```

```python
#Make the points give more details when you hover above it

x = 'confidence when correct'
y = 'confidence when incorrect'

df_insomnia=df.loc[df['insomnia'], :]
df_normal = df.loc[~df['insomnia'], :]

p = bokeh.plotting.figure(
    frame_width = 300,
    frame_height = 300,
    x_axis_label = x,
    y_axis_label = y,
    x_range = [-5,105],
    y_range = [-5,105],
    tooltips=[ # so hovering gives more info
        ('p-number', '@{participant number}'),
        ('gender', '@gender'),
        ('age', '@age')
    ]
)

source = bokeh.models.ColumnDataSource(df) #notneeded

p.line(
    x=[-5, 105],
    y=[-5, 105],
    color='gray'
)

p.circle(
    source=df_normal,
    x=x,
    y=y,
    #color = 'blue',
    legend_label='normal'
)

p.circle(
    source=df_insomnia,
    x=x,
    y=y,
    color = 'orange',
    legend_label='insomniacs'
)

p.legend.location = 'top_left'
p.legend.click_policy = 'mute'

p.output_backend='svg' #so download is svg, not png
bokeh.io.show(p)
```

```python
bokeh.io.save(
    p,
    filename='our_first_bokeh_plot_together.html',
    title='GFMT Confidence',
    resources=bokeh.resources.CDN
)
```
